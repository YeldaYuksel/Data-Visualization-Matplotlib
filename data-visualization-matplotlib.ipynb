```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#df = sns.load_dataset("titanic")
df = pd.read_csv("/kaggle/input/titanic-dataset-csv/titanic.csv")
df.head()
```

```python
##Visualization with barplot

df["sex"].value_counts().plot(kind='bar')
plt.show()
```

```python
##Visualization with hist

plt.hist(df["age"])
plt.show()
```

```python
##Visualization with boxplot

(plt.boxplot(df["fare"]))
plt.show()
```

**Plot**

```python
x = np.array([1, 8])
y = np.array([0, 150])
plt.plot(x, y)
plt.show()
```

```python
x = np.array([1, 8])
y = np.array([0, 150])
plt.plot(x, y, 'o')
plt.show()
```

**Marker**

```python
x = np.array([1, 8])
y = np.array([0, 150])
plt.plot(x, y, marker='*')
plt.show()
```

**Line**

```python
x = np.array([1, 8])
y = np.array([0, 150])
plt.plot(x, y, linestyle="dotted")
plt.show()
```

**Multiple Lines**

```python
x = np.array([1, 8])
y = np.array([0, 150])
plt.plot(x)
plt.plot(y)
plt.show()
```
